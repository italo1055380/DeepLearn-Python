## **Desafio 01 Deep Learning - Reperesentação dos Dados**

```python
# Importação da biblioteca
import numpy as np

# A função do codigo abaixo é carregar e ler o conteudo dos arquivos dos textos: 
# Base Text - Matrix Representation I.txt
# Base Text - Matrix Representation II.txt

with open('/content/drive/MyDrive/Base Text - Matrix Representation I.txt', 'r') as file1, open('/content/drive/MyDrive/Base Text - Matrix Representation II.txt', 'r') as file2:
    file_content1 = file1.read()
    file_content2 = file2.read()
    
   
# Esta parte do código foi tradado a limpeza dos codigo como Remover vírgula, espaço, ponto, aspas e ponto e vírgula usando a função replace
    
    file_content1 = file_content1.replace(',', '').replace(' ', '').replace('.', '').replace('"', '').replace(';', '').replace(':', '').replace("'", '').replace('"', '').replace('(', '').replace(')', '')
    file_content2 = file_content2.replace(',', '').replace(' ', '').replace('.', '').replace('"', '').replace(';', '').replace(':', '').replace("'", '').replace('"', '').replace('(', '').replace(')', '')

    
# Apos a remoção algumas strings foi concatenado os dois textos no combined_content para que podesemos aplicar na Matriz unica de dados.
    combined_content = file_content1 + file_content2

# Criação da Matriz de dados já usando o combined_content e por fim imprimindo os resultado.
    matrix = []
    for line in combined_content.split('\n'):
        row = line.split()
        matrix.append(row)

    # Imprimir a matriz
    for row in matrix:
        print(row)



```

```python
# Para converter a matriz de volta ao texto original, 
# você precisará percorrer cada linha e combinar
# os elementos de cada linha em uma única string. 
# Em seguida, você pode concatenar todas as strings de linha em um único texto
# usando um caractere de nova linha como separador.

# Percorrer cada linha da matriz e combinar os elementos em uma única string
lines = []
for row in matrix:
    line = ' '.join(row)
    lines.append(line)

# Concatenar as strings de linha em um único texto
text = '\n'.join(lines)

# Imprimir o texto
print(text)



```
