## **Desafio 01 Deep Learning - Reperesentação dos Dados**

```python
# Importação da biblioteca
import re

# A função do codigo abaixo é carregar e ler o conteudo do arquivo de texto: 
# Base Text - Matrix Representation I.txt

with open('/content/drive/MyDrive/Base Text - Matrix Representation I.txt', 'r') as file1:
    file_content1 = file1.read()
    print(file_content1)
```

```python
# Pré-processamento o tratamento dos dados antes do processaento de fato(retirda de pntuações e caracteres especiais)
clean_text = re.sub('[.,;:"\'()\r!@#$%^&*\u2029]', '', file_content1)
```

```python
# Criação da Matriz de dados já usando o filecontent1 e por fim imprimindo os resultado.
    
matrix = []
for line in file_content1.split(','):
        row = line.split()
        matrix.append(row)

# Imprimir a matriz
for row in matrix:
        print(row)
```

```python
# Para converter a matriz de volta ao texto original, 
# você precisará percorrer cada linha e combinar
# os elementos de cada linha em uma única string. 
# Em seguida, você pode concatenar todas as strings de linha em um único texto
# usando um caractere de nova linha como separador.

# Percorrer cada linha da matriz e combinar os elementos em uma única string
lines = []
for row in matrix:
    line = ' '.join(row)
    lines.append(line)

# Concatenar as strings de linha em um único texto
text = '\n'.join(lines)

# Imprimir o texto
print(text)
```
