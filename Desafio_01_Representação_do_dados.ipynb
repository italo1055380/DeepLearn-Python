```python

# Ler o conteúdo dos arquivos de texto e concatená-los em uma única string
with open('/content/drive/MyDrive/Base Text - Matrix Representation I.txt', 'r') as file1, open('/content/drive/MyDrive/Base Text - Matrix Representation II.txt', 'r') as file2:
    file_content1 = file1.read()
    file_content2 = file2.read()
    combined_content = file_content1 + file_content2

# Separar as palavras distintas da string e contar sua ocorrência
words = combined_content.split()
word_count = collections.Counter(words)

# Criar a matriz a partir das palavras distintas e seu total de ocorrências
word_list = sorted(set(words))
matrix = np.zeros((len(word_list), len(words)))
for i, word in enumerate(word_list):
    for j, text_word in enumerate(words):
        if word == text_word:
            matrix[i][j] = word_count[word]

# Imprimir a matriz

print(matrix)

```
